```python
from pydantic import BaseModel, Field
from typing import Optional, Literal


class TodoItem(BaseModel):
    id: str = Field(description="Unique item identifier")
    content: str = Field(description="Task description")
    status: Literal["pending", "in_progress", "completed", "cancelled"] = Field(
        description="Current status"
    )


class TodoWriteToolInput(BaseModel):
    todos: Optional[list[TodoItem]] = Field(
        default=None,
        description="Task items to write. Omit to read current list.",
    )
    merge: bool = Field(
        default=False,
        description=(
            "true: update existing items by id, add new ones. "
            "false (default): replace the entire list."
        ),
    )

from pydantic.json_schema import GenerateJsonSchema


class InlineJsonSchema(GenerateJsonSchema):
    def ref_schema(self, json_schema):
        # 强制展开引用，不生成 $ref
        return self.generate_inner(json_schema)
```

```python
todo = TodoWriteToolInput(item="Buy milk", checked=False, path="TODO.md").model_json_schema(mode="serialization", schema_generator=InlineJsonSchema)
from pprint import pprint
pprint(todo)
```
