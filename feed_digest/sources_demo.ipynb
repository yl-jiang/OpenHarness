# Feed Digest Sources Demo

运行 `sources.py` 中每个 source 的 `collect()`，展示原始返回结果。

```python
import sys
from pathlib import Path

# Ensure repo root is on PYTHONPATH so `feed_digest` is importable
repo_root = Path().resolve().parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

import asyncio
from datetime import datetime, timezone, timedelta

from feed_digest.sources import (
    GitHubTrendingSource,
    HackerNewsSource,
    RSSSource,
    V2EXSource,
    HuggingFacePapersSource,
    ArXivSource,
    Kr36Source,
    ITHomeSource,
    NewsNowSource,
    SoPilotSource,
    BrowserCommandSource,
)
from feed_digest.models import FeedItem


def items_to_dicts(items: list[FeedItem]) -> list[dict]:
    """Convert FeedItem dataclass instances to plain dicts for display."""
    return [item.__dict__ for item in items]


since = (datetime.now(timezone.utc) - timedelta(days=7)).isoformat()
until = datetime.now(timezone.utc).isoformat()
domain = "ai"
query = "AI"
max_items = 5

print(f"since={since}")
print(f"until={until}")
print(f"domain={domain!r}")
print(f"query={query!r}")
print(f"max_items={max_items}")
```

## 1. GitHubTrendingSource

```python
source = GitHubTrendingSource()
items = await source.collect(since=since, until=until, domain=domain, query=query, max_items=max_items)
items_to_dicts(items)
```

## 2. HackerNewsSource (search mode)

```python
source = HackerNewsSource(mode="search")
items = await source.collect(since=since, until=until, domain=domain, query=query, max_items=max_items)
items_to_dicts(items)
```

## 批量汇总

```python
sources = [
    ("GitHubTrending", GitHubTrendingSource()),
    ("HackerNews", HackerNewsSource()),
    ("RSS", RSSSource()),
    ("V2EX", V2EXSource()),
    ("HuggingFacePapers", HuggingFacePapersSource()),
    ("ArXiv", ArXivSource()),
    ("Kr36", Kr36Source()),
    ("ITHome", ITHomeSource()),
    ("NewsNow", NewsNowSource()),
    ("SoPilot", SoPilotSource()),
    ("BrowserCommand", BrowserCommandSource()),
]

summary = {}
for name, src in sources:
    try:
        items = await src.collect(since=since, until=until, domain=domain, query=query, max_items=max_items)
        summary[name] = len(items)
    except Exception as exc:
        summary[name] = f"ERROR: {exc}"

summary
```
